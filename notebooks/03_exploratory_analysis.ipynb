# Exploratory analysis

This notebook documents the DuckDB analysis used by the app. It analyzes only complete annual baskets and does not fill missing observations.

```python
import duckdb
con = duckdb.connect()
con.execute("CREATE TABLE prices AS SELECT * FROM read_csv_auto('../data/processed/annual_grocery_prices.csv', header=True)")
con.sql("SELECT year, ROUND(SUM(annual_average_price * basket_quantity), 2) AS basket_cost FROM prices GROUP BY year HAVING COUNT(DISTINCT series_id) = 8 ORDER BY year")
```
